```python
import pandas as pd
```

# Import datasets and validate

```python
path_failures = '../dataset/wind-farm-1-failures-training.csv'
path_dataset = '../dataset/wind-farm-1-signals-training.json'
```

## dataset

```python
df = pd.read_json(path_dataset)
df = pd.DataFrame.from_dict(df['fields'].tolist())
df.head()
```

```python
# make all columns as lowercase
df.columns = [col.lower() for col in df.columns]

# make timestamp, turbine_id as first columns
cols = df.columns.tolist()
first_cols = ['timestamp','turbine_id']
[cols.remove(c) for c in first_cols]
cols = first_cols + cols
df = df[cols].copy()

# make timestamp as datetime type
df['timestamp'] = pd.to_datetime(df['timestamp'])
print(f'Dimensions of the dataset: {df.shape}')
df.head(10)
```

```python
df.info()
```

## failures

```python
failures = pd.read_csv(path_failures, sep=';')
failures.columns = [col.lower() for col in failures.columns]
failures['timestamp'] = pd.to_datetime(failures['timestamp'])
print(f'Dimensions of the dataset: {failures.shape}')
failures
```

```python
failures.info()
```

## merging datasets

```python
df_merged = df.merge(failures, how='outer', on=['timestamp', 'turbine_id'])
```

```python
# add failure column
df_merged.loc[:,['failure']] = 0
df_merged.loc[~df_merged['component'].isnull(),['failure']] = 1
```

## Splitting the datasets for each turbine

```python
for turbine in df_merged['turbine_id'].unique():
    df_turbine = df_merged.loc[(df_merged['turbine_id'] == turbine),:]
    df_turbine.to_json(f'../dataset/df_turbine_{turbine}.json') 
    print(f'************ Dataset saved for turbine {turbine}  ************')
    
```
